```python
# import dependencies, import file, create DF
import pandas as pd
file_to_read = "vine_table.csv"
vine_df = pd.read_csv(file_to_read)
```

```python
# Check that it imported (for my own sanity)
vine_df
```

```python
#Filter to find were there are more than 20 total votes
vine_df = vine_df.loc[vine_df["total_votes"] > 20]
vine_df
```

```python
# Create new dataframe where helpful / total >= 50%
helpful_df = vine_df.loc[vine_df['helpful_votes'] / vine_df['total_votes'] >= .5]
helpful_df
```

```python
paid_helpful_df = helpful_df.loc[helpful_df['vine'] == "Y"]
paid_count_total = paid_helpful_df['total_votes'].count()
paid_5star_df = paid_helpful_df.loc[paid_helpful_df['star_rating'] == 5]
paid_5star = paid_5star_df['star_rating'].count()
paid_percentage = ((paid_5star / paid_count_total) *100).round()
print(f"Vine Total Votes: {paid_count_total}")
print(f"Vine 5 Star Reviews: {paid_5star}")
print(f"Vine Percentage: {paid_percentage}%")
```

```python
unpaid_helpful_df = helpful_df.loc[helpful_df['vine'] == "N"]
unpaid_count_total = unpaid_helpful_df['total_votes'].count()
unpaid_5star_df = unpaid_helpful_df.loc[unpaid_helpful_df['star_rating'] == 5]
unpaid_5star = unpaid_5star_df['star_rating'].count()
unpaid_percentage = ((unpaid_5star / unpaid_count_total) *100).round()
print(f"Non-Vine Total Votes: {unpaid_count_total}")
print(f"Non-Vine 5 Star Reviews: {unpaid_5star}")
print(f"Non-Vine Percentage: {unpaid_percentage}%")
```
